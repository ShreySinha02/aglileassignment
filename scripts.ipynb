```python
import pandas as pd
```

```python
df=pd.read_csv('User IDs.csv')
df1=pd.read_csv('Rigorbuilder RAW.csv')
```

```python
df.rename(columns = {'User ID':'uid'}, inplace = True)
df
```

```python
df1
```

```python
df2= pd.merge(df, df1, how = 'inner',left_on=['uid'],
                  right_on=['uid'],suffixes=('', '_remove'))
```

```python
df2=df2.drop(['S No_remove','name'],axis=1)
```

```python
df2['Team Name'].replace(['BrandTech Lab','BrandTech Lab\t'],'BrandTech Lab',inplace=True)
```

```python
df3 =df2.groupby(['Team Name'])[['total_statements','total_reasons']].mean()
```

```python
df4=df3.sort_values(by=['total_statements'],ascending=False)
```

```python
df4.rename(columns={'Team Name':'Thinking Teams Leaderboard','total_statements':'Average Statements','total_reasons':'Average reasons'},inplace=True)
```
